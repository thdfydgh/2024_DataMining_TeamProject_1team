```python
import pandas as pd

df = pd.read_csv("2022서울시교통사고현황-데이터셋.csv",encoding = 'utf-8')
```

```python
df.columns
```

```python
df1 = df[df["사고유형 - 대분류"] == "차량단독 "]  # 차량단독
df2 = df[df["사고유형 - 대분류"] == "차대차 "]  # 차대차
df3 = df[df["사고유형 - 대분류"] == "차대사람 "]  # 차대사람
```

```python
df3
```

## 차대차

```python
feature2_counts = df2.groupby('행정동')['사고유형 - 소분류'].value_counts().unstack(fill_value=0)
feature2_counts.columns = [f'사고유형 - 소분류_{col}' for col in feature2_counts.columns]
```

```python
feature3_counts = df2.groupby('행정동')['법규위반'].value_counts().unstack(fill_value=0)
feature3_counts.columns = [f'법규위반_{col}' for col in feature3_counts.columns]
```

```python
feature4_counts = df2.groupby('행정동')['도로형태 - 대분류'].value_counts().unstack(fill_value=0)
feature4_counts.columns = [f'도로형태 - 대분류_{col}' for col in feature4_counts.columns]
```

```python
feature5_counts = df2.groupby('행정동')['도로형태 - 소분류'].value_counts().unstack(fill_value=0)
feature5_counts.columns = [f'도로형태 - 소분류_{col}' for col in feature5_counts.columns]
```

```python
feature6_counts = df2.groupby('행정동')['가해운전자 차종'].value_counts().unstack(fill_value=0)
feature6_counts.columns = [f'가해운전자 차종_{col}' for col in feature6_counts.columns]
```

```python
feature7_counts = df2.groupby('행정동')['가해운전자 연령대'].value_counts().unstack(fill_value=0)
feature7_counts.columns = [f'가해운전자 연령대_{col}' for col in feature7_counts.columns]
```

```python
aggregated_df2 = df2.groupby('행정동').sum().reset_index()
```

```python
aggregated_df2.head()
```

```python
df2_counts = df2[['행정동']].drop_duplicates().set_index('행정동')
df2_counts = df2_counts.join(feature2_counts).join(feature3_counts).join(feature4_counts).join(feature5_counts).join(feature6_counts).join(feature7_counts).reset_index()
```

```python
df2_counts.head()
```

```python
df2_counts['ECLO'] = aggregated_df2['ECLO']
```

```python
df2_counts.head()
```

```python
count_df2 = df2.groupby('행정동').size().reset_index(name='행정동_개수')
```

```python
final_df2 = pd.merge(df2_counts, count_df2, on='행정동')
```

```python
final_df2.head()
```

```python
final_df2['평균 ECLO'] = final_df2['ECLO'] / final_df2['행정동_개수']
```

```python
final_df2.head()
```

```python
csv_file_path = '2022_final_차대차.csv'
final_df2.to_csv(csv_file_path, index=False, encoding='cp949')
```

## 차대 사람

```python
feature2_counts = df3.groupby('행정동')['사고유형 - 소분류'].value_counts().unstack(fill_value=0)
feature2_counts.columns = [f'사고유형 - 소분류_{col}' for col in feature2_counts.columns]

feature3_counts = df3.groupby('행정동')['법규위반'].value_counts().unstack(fill_value=0)
feature3_counts.columns = [f'법규위반_{col}' for col in feature3_counts.columns]

feature4_counts = df3.groupby('행정동')['도로형태 - 대분류'].value_counts().unstack(fill_value=0)
feature4_counts.columns = [f'도로형태 - 대분류_{col}' for col in feature4_counts.columns]

feature5_counts = df3.groupby('행정동')['도로형태 - 소분류'].value_counts().unstack(fill_value=0)
feature5_counts.columns = [f'도로형태 - 소분류_{col}' for col in feature5_counts.columns]

feature6_counts = df3.groupby('행정동')['가해운전자 차종'].value_counts().unstack(fill_value=0)
feature6_counts.columns = [f'가해운전자 차종_{col}' for col in feature6_counts.columns]

feature7_counts = df3.groupby('행정동')['가해운전자 연령대'].value_counts().unstack(fill_value=0)
feature7_counts.columns = [f'가해운전자 연령대_{col}' for col in feature7_counts.columns]

aggregated_df3 = df3.groupby('행정동').sum().reset_index()

df3_counts = df3[['행정동']].drop_duplicates().set_index('행정동')
df3_counts = df3_counts.join(feature2_counts).join(feature3_counts).join(feature4_counts).join(feature5_counts).join(feature6_counts).join(feature7_counts).reset_index()


df3_counts['ECLO'] = aggregated_df3['ECLO']

df3_counts.head()

count_df3 = df3.groupby('행정동').size().reset_index(name='행정동_개수')

final_df3 = pd.merge(df3_counts, count_df3, on='행정동')

final_df3.head()

final_df3['평균 ECLO'] = final_df3['ECLO'] / final_df3['행정동_개수']

final_df3.head()

csv_file_path = '2022_final_차대사람.csv'
final_df3.to_csv(csv_file_path, index=False, encoding='cp949')
```
